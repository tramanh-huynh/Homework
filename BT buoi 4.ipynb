# Task 4: In ra các thành phố có dân số & diện tích đều nằm trong Top 20

```python
#print cities having population and area are both in top 20
top_20 = df.loc[:,["City","Population","Area KM2"]]
top_20["Population"] = top_20["Population"].apply(lambda x: x.replace(",",""))
top_20["Area KM2"] = top_20["Area KM2"].apply(lambda x: x.replace("-May",''))
top_20[["Population","Area KM2"]] = top_20[["Population","Area KM2"]].astype(int)
```

```python
top_20_pop = top_20.loc[:,["City","Population"]]
top_20_pop = top_20_pop.sort_values("Population",ascending=False)
top_20_pop1 = top_20_pop[:20]
top_20_area = top_20.loc[:,["City","Area KM2"]]
top_20_area = top_20_area.sort_values("Area KM2", ascending=False)
top_20_area1 = top_20_area[:20]
```

```python
cities_both = top_20_pop1.merge(top_20_area1, on='City')
cities_both
```

# Task 5: Thống kê mật độ dân số theo quốc gia

```python
pop_density = df.loc[:,["Country","Population","Area KM2"]]
pop_density["Population"] = pop_density["Population"].apply(lambda x: x.replace(",",""))
pop_density["Area KM2"] = pop_density["Area KM2"].apply(lambda x: x.replace("-May",''))
pop_density["Country"] = pop_density["Country"].apply(lambda x: x.replace(" *",''))
pop_density[["Population","Area KM2"]] = pop_density[["Population","Area KM2"]].astype(int)
pop_density = pop_density.groupby(["Country"]).sum()
#pop_density["Country"].value_counts()
pop_density
```

```python
pop_density["Population Density"] = pop_density["Population"]/pop_density["Area KM2"]
pop_density
```

# Task 6: Thống ke các thành phố có dân số lớn nhất của từng quốc gia (chỉ tính của những quốc gia có 2 thành phố xuất hiện trở lên trong bảng)

```python
pop_city = df.loc[:,["City","Population","Country"]]
pop_city["Population"] = pop_city["Population"].apply(lambda x: x.replace(",",""))
pop_city["Country"] = pop_city["Country"].apply(lambda x: x.replace(" *",''))
pop_city[["Population"]] = pop_city[["Population"]].astype(int)

country_count = pop_city["Country"].value_counts()
```

```python
condition = country_count[country_count > 1]
pop_city2 = pop_city[pop_city["Country"].isin(condition.index)]
pop_city2["Country"].value_counts()
```

```python
top_2_cities = pop_city2.groupby(by="Country")["City","Population"].apply(lambda x: x.nlargest(1, "Population"))
top_2_cities
```
