## Task 1 : Job title nào có mức lương trung bình theo USD cao nhất ?

## Task 9: Job nào có khoảng cách thu nhập giữa SE và MI ít nhất ?

```python
task9 = data[(data["experience_level"]!="EN")&
            (data["experience_level"]!="EX")][["job_title","experience_level","salary_in_usd"]]
task9.head()
```

```python
task9 = task9.groupby(["job_title","experience_level"]).mean()
task9
```

```python
#xoa nhung job title co 1 dong vi ko the so sanh
task9_1 = task9.groupby(level=0).size()
task9_2 = task9_1[task9_1 > 1].index
task9_3 = task9.loc[task9_2]
task9_3
```

```python
#index co 2 level : slice(None) = chon tat ca gia tri cua 1st index
#2nd index lay gia tri SE/MI
SE = task9_3.loc[(slice(None), 'SE'), :]
MI = task9_3.loc[(slice(None), 'MI'), :]
```

```python
#SE
#MI
```

```python
distance = SE.loc[:,"salary_in_usd"].values - MI.loc[:,"salary_in_usd"].values
distance
```

```python
index = task9_3.index
job = [idx[0] for idx in index]
job = set(job)
```

```python
merge = list(zip(job,distance))
merge
```

## Task 10: Job nào có mức lương trung bình đều nằm trong top 5 ở tất cả các level ?

```python
task10 = data[["job_title","experience_level","salary_in_usd"]]
task10_1 = task10.groupby(["experience_level","job_title"]).mean()
task10_1.head()
```

```python
top_5 = task10_1.groupby("experience_level").apply(lambda x: x.sort_values("salary_in_usd", ascending=False).head(5))
top_5.head()
```

```python
top_5_job = top_5.index
top_5_job = [x[1:] for x in top_5_job]
top_5_job
```

```python
job_dict = {}
for exp_level, job in top_5_job:
    if job not in job_dict:
        job_dict[job] = set()
    job_dict[job].add(exp_level)

job_dict
```
